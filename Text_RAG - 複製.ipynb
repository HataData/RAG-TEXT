```python
import os
from dotenv import load_dotenv
from langchain_community.document_loaders import PyMuPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_google_genai import GoogleGenerativeAIEmbeddings 
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS

from tenacity import retry, stop_after_attempt, wait_random_exponential
@retry(wait=wait_random_exponential(min=1, max=60), stop=stop_after_attempt(5))
def safe_vectorize():
    print("loading...")
    return FAISS.from_documents(documents=splits, embedding=embeddings)

try:
    vectorstore = safe_vectorize()
    print("createdb")
except Exception as e:
    print(f"check network {e}")
    
embeddings = GoogleGenerativeAIEmbeddings(
    model="text-embedding-004",
    # timeout setting
    request_options={'timeout': 60} 
)
from langchain_community.vectorstores import FAISS


# 
current_dir = os.getcwd()
print(f"current directory {current_dir}")

# 2. read.env file
# override=True cove memorry before
load_dotenv(dotenv_path=os.path.join(current_dir, ".env"), override=True)


api_key = os.getenv("GOOGLE_API_KEY")
# 3. check
if api_key:
    # 
    api_key = api_key.strip().replace('"', '').replace("'", "")
    # 
    os.environ["GOOGLE_API_KEY"] = api_key
    
    print(f"success: {api_key[:10]}...")
else:
    print("can't find GOOGLE_API_KEY。")
   
# 2. file path
pdf_path = "data/2306.08045v2.pdf" 

# --- step1 slice---
loader = PyMuPDFLoader(pdf_path)
docs = loader.load()
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=150)
splits = text_splitter.split_documents(docs)
print(f"read {len(docs)} page，slice {len(splits)} document。")

# --- use model ---
# "models/text-embedding-004"
#embeddings = GoogleGenerativeAIEmbeddings(model="models/text-embedding-004")
embeddings = GoogleGenerativeAIEmbeddings(model="text-embedding-004")

# create faiss db
vectorstore = FAISS.from_documents(documents=splits, embedding=embeddings)

print("created db success（ Google Embedding）。")

# --- search check---
query = "What are the main contributions of this paper?"
search_results = vectorstore.similarity_search(query, k=2)

print("\n🔍 search result：")
for i, result in enumerate(search_results):
    print(f"\n--- from {i+1} ---")
    print(result.page_content[:200] + "...")
```
